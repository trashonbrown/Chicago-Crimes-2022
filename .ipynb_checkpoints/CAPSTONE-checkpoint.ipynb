# **Crimes in Chicago (2022)**

## *1) Read in Data*

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# read the csv file into a DataFrame
crime = pd.read_csv('2022.csv')

crime.info()
```

## *2) Hypothesis*

"The rate of theft crimes is higher in the downtown district than in the other districts of Chicago."

This hypothesis suggests that there is a relationship between the type of crime and the district where the crime occurred. The hypothesis predicts that the rate of theft crimes will be higher in the downtown district than in the other districts of Chicago.



## *3) Finding the Total Number of Missing Data*

```python
crime.isnull().sum()
```

## *4) Removing the Missing Data*

The reasoning behind removing the location-based columns are because the identity and addresses of the victims cannot be disclosed. To test the hypothesis, I created a new "Location" column that specifies whether a crime has been committed in a district that is either in the Downtown area or outside of the Downtown area.

```python
crime = crime.drop(['Latitude', 'Longitude', 
                'X Coordinate', 'Y Coordinate', 'Location'], axis=1)
```

## *5) Only Consider Theft Crimes*

```python
crime = crime[crime["Primary Type"] == "THEFT"]
crime
```

## *6) Group the Data by District and Calculate the Number of Theft Crimes*


```python
theft_by_district = crime.groupby("District").size().reset_index(name="Theft Count")
theft_by_district
```

## *7) Visualize Step #6*

```python
theft_by_district.plot.barh(x='District', y= 'Theft Count')
plt.title("Theft Count by District")
plt.xlabel("Theft Count")
plt.ylabel("Districts")
plt.show()
```

## *8) Create a Boolean Column for if the District is Downtown*

```python
crime["Location"] = crime["District"].apply(lambda x: "Downtown" if x == 1 else "Other")
crime
```

```python
crime.loc[crime['District'] == True]
```

## *9) Group the Data by if the District is Downtown and Calculate the Mean Number of Thefts*

```python
theft_rate = crime.groupby("Location").size().reset_index(name="Theft Count")
theft_rate
```

## *10) Plot the Theft Rate by District*

```python
plt.bar(theft_rate["Location"], theft_rate["Theft Count"])
plt.xlabel("Location")
plt.ylabel("Theft Count")
plt.show()
```

## *11) Perform a Chi-squared Test to Determine if the Rate of Thefts in Downtown is Significantly Higher than in Other Districts*

```python
contingency_table = pd.crosstab(theft_rate["Theft Count"], [theft_rate['Location']])
contingency_table
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)
```

## *12) Check the P-Value*


```python
if p < 0.05:
    print("The rate of thefts in downtown is significantly higher than in other districts")
else:
    print("The rate of thefts in downtown is not significantly higher than in other districts")
```

The hypothesis "The rate of theft crimes is higher in the downtown district than in the other districts of Chicago" was tested using an exploratory data analysis of Chicago's 2022 crime database.  After cleaning and analyzing the data, it was found that the rate of theft crimes is not significantly higher in the downtown district than in the other districts of Chicago.
